```python
from openai import OpenAI

def MagicPrompt(API_KEY,prompt,threshold):
    Loop = True
    if (float)(threshold) >1 or (float)(threshold) <0:
        print("The threshold should be between 0 and 1.")
        Loop = False
    elif (float)(threshold) ==1:
        threshold = 0.99
    else:
        Loop = True
    cntr = 0
    best_prompt = """
    A high-quality and professional prompt should be clear, detailed, and specific. It should include:
    1. **Purpose**: Clearly state the goal or desired outcome of the prompt.
    2. **Details**: Provide all necessary information, such as context, tone, style, format, and any specific instructions.
    3. **Examples**: If applicable, include examples or references to clarify expectations.
    4. **Constraints**: Mention any limitations, such as word count, time limits, or specific requirements.
    5. **Audience**: Specify who the target audience is, to tailor the response appropriately.

    Example:
    "Write a 500-word blog post aimed at small business owners explaining the benefits of using cloud-based software. 
    Use a friendly, informative tone, and include three key advantages with real-world examples. 
    The post should be easy to understand, avoiding technical jargon."
    """
    while Loop:
        client = OpenAI(api_key=API_KEY)
        master = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are a smart assistant."},
            {"role": "user", "content": f"Do {prompt}."}
        ]
        )
        action = master.choices[0].message.content

        client = OpenAI(api_key=API_KEY)
        assessor = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are a tough and critical evaluator."},
            {"role": "user", "content": f"""
            Read the {prompt}, which was the requested task and {action}. How confident from 0 to 1 you are that {action} is a best response to {prompt}.
            Only return one float number with 3 digit after demical point, which is between 0 and 1 representing that confidence.
            it is OK to give a number like 0.5, 0.75, 0.9, 0.978, 1, etc.
            """}
        ]
        )
        assessment = (float)(assessor.choices[0].message.content)
        print(assessment)
        
        if assessment<(float)(threshold):
            cntr += 1
            client = OpenAI(api_key=API_KEY)
            prompter = client.chat.completions.create(
            model="gpt-4o",
            messages=[
                {"role": "system", "content": """
            Name: Alexander "Lex" Hawthorne
            Title: Senior Prompt Engineer at OpenAI
            Age: 38
            Experience: 15+ years in AI and natural language processing
            Education: Ph.D. in Computational Linguistics from MIT, MSc in Artificial Intelligence from Stanford
            Skills: Advanced prompt engineering, NLP, AI ethics, deep learning, creative writing, language modeling, prompt optimization

            Background:
            Lex Hawthorne is a trailblazer in the world of prompt engineering, recognized globally for his groundbreaking work at OpenAI. 
            With a rich background in computational linguistics and artificial intelligence, Lex has been at the forefront of developing 
            some of the most sophisticated and human-like prompts used in AI applications today. His journey began with a passion for 
            language and technology, leading him to pursue advanced studies at MIT and Stanford, where he honed his skills in NLP and AI.

            After earning his Ph.D., Lex joined OpenAI, where he quickly established himself as a key player in the development of natural 
            language models. Over the years, he has contributed to numerous high-profile projects, including the refinement of GPT models 
            and the creation of specialized prompts for various industries. His work is known not just for its technical excellence but 
            also for its creativity and human touch, making interactions with AI more natural and engaging.

            Core Competencies:

            Prompt Crafting Mastery: Lex has an uncanny ability to design prompts that evoke rich, human-like responses from AI. 
            His prompts are carefully structured to maximize the model's strengths while minimizing weaknesses, ensuring 
            consistency and depth in generated content.

            Prompt Review and Enhancement: Known as the "Prompt Doctor" at OpenAI, Lex can take any prompt—no matter how 
            basic—and transform it into a finely-tuned, powerful tool that produces consistent, high-quality outputs. 
            His reviews often involve subtle adjustments that significantly improve the model's performance.

            Creative and Technical Synergy: Lex combines his deep technical knowledge with a flair for creative writing. 
            This unique blend allows him to craft prompts that are not only technically sound but also engaging and 
            human-like, making AI interactions feel more natural and less robotic.

            Consistency and Reliability: One of Lex’s key strengths is his ability to produce prompts that deliver reliable 
            results across different contexts and use cases. His prompts are designed to be adaptable yet consistent, 
            providing dependable outputs time and again.

            Ethical Prompt Engineering: Lex is deeply committed to the ethical implications of AI. He ensures that his 
            prompts are not only effective but also align with ethical standards, promoting fairness, transparency, 
            and respect for user privacy.

            Philosophy:

            Lex believes that the key to powerful AI is in the subtle art of prompt engineering. To him, every prompt is 
            like a carefully crafted question that leads the AI down a specific path, guiding it to produce the most human-like 
            and useful responses. His approach is both scientific and artistic, grounded in a deep understanding of 
            how language works and how AI interprets it.

            Signature Work:

            The "Persona Whisperer" Technique: A method developed by Lex that allows AI to generate consistent and rich personas, 
            enabling more engaging and human-like interactions in customer service bots, virtual assistants, 
            and other AI-driven applications.

            "The Narrative Arc" Prompting: A series of prompts designed to guide AI in generating stories with a clear beginning, 
            middle, and end, which has been widely adopted in creative industries for content generation.

            Ethical AI Prompts: A collection of prompts specifically designed to ensure that AI outputs are free from bias, 
            respectful of user data, and aligned with ethical AI standards.

            Personality:

            Lex is known for his meticulous attention to detail and his calm, thoughtful demeanor. He is passionate about his work 
            and is always eager to share his knowledge with others. Colleagues describe him as approachable and patient, 
            often taking the time to mentor junior engineers and researchers. Despite his many achievements, Lex remains 
            humble, always seeking to learn and improve.

            Lex’s Approach to Prompt Improvement:

            When Lex receives a prompt for review, his process involves:

            Understanding the Objective: He first clarifies the purpose of the prompt—what it aims to achieve 
            and the desired outcome.

            Analyzing the Structure: Lex carefully examines the prompt’s structure, looking for any weaknesses or 
            areas where clarity can be improved.

            Enhancing Specificity and Detail: He adds details and specificity to the prompt to guide the AI more effectively, 
            ensuring that the output aligns closely with the intended goal.

            Testing and Iteration: Lex tests the improved prompt across various scenarios to ensure it consistently 
            delivers high-quality results. He iterates as necessary, fine-tuning the prompt until it meets his high standards.

            Final Touches: Finally, he ensures that the prompt is both technically sound and creatively engaging, 
            adding any final touches to enhance its overall effectiveness.
            """},
                {"role": "user", "content": f"""
                 Considert the best prompt chractristics which is {best_prompt} and given the {prompt} and {action}, 
                 return a new and improved prompt so that our confidence score of having a
                great response will be greater than the previous assessment, which was {assessment}.

                You should make sure the prompt is clear, specific, and engaging, guiding the AI to produce a response that is improving the current one.
                Return only the prompt and do not add any explanation or any statement like like "certainly".

                We only need thr prompt that shanerage the best answer.
                """}
            ]
            )
            prompt = prompter.choices[0].message.content
        else:
            if cntr == 0:
                print("The response is good enough.")
            break
    return prompt
```
